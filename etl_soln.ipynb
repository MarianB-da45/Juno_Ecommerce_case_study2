# Case_Study

```python
url = 'https://www.jumia.com.ng/catalog/?q=laptops'
r = requests.get(url)


soup = BeautifulSoup(r.text, "lxml")
```

```python
# products names
product_names =[]
names = soup.find_all('h3', class_ = "name")
for name in names:
        product_names.append(name.string)
```

```python
## current price
current_prices =[]
prices = soup.find_all('div', class_ = "prc")
for price in prices:
        current_prices.append(price.string)
```

```python
## former price
former_prices =[]
prices = soup.find_all('div', class_ = "old")
for price in prices:
        former_prices.append(price.string)
```

```python
## discount percentages
discount_percs =[]
discounts = soup.find_all('div', class_ = "bdg _dsct _sm")
for discount in discounts:
    discount_percs.append(discount.string)
```

```python

data = {
    'title': product_names, 
    'current_price': current_prices, 
    'former_price': former_prices,
    'discount': discount_percs
    }
laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
```

```python
laptop_df = laptop_df.transpose()
laptop_df
```

```python
# for all the pages 
pages = [x for x in range(1,51)]
product_names =[]
current_prices =[]
former_prices =[]
discount_percs =[]

for page in pages:
    url = f'https://www.jumia.com.ng/catalog/?q=laptops&page={page}#catalog-listing'
    
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")

    # products names
    names = soup.find_all('h3', class_ = "name")
    for name in names:
        product_names.append(name.string)
        ## current price
    prices = soup.find_all('div', class_ = "prc")
    for price in prices:
        current_prices.append(price.string)
    ## former price
    prices = soup.find_all('div', class_ = "old")
    for price in prices:
        former_prices.append(price.string)
     ## discount percentages
    discounts = soup.find_all('div', class_ = "bdg _dsct _sm")
    for discount in discounts:
        discount_percs.append(discount.string)

data = {
    'title': product_names, 
    'current_price': current_prices, 
    'former_price': former_prices,
    'discount': discount_percs
    }
laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
laptop_df = laptop_df.transpose()
laptop_df
```

```python
laptop_df.fillna({
    'former_price': '0',
    'discount': '0%'
}, inplace=True)
```

```python
# Database credentials
username = os.getenv('username')
password = os.getenv('password')
host = os.getenv('host')
port = os.getenv('port')
database = os.getenv('database')
```

```python
# define the database connection parameters including the database name


# define the database connection url
db_url = f"postgresql://{username}:{password}@{host}:{port}/{database}"

# create the database engine with the db url specified
engine = create_engine(db_url)

# connect to postgreSQL server
with engine.connect() as connection:
    
  # create table and load the data
  laptop_df.to_sql('Juno_Ecommerce_products', connection, index=False, if_exists='replace')

print('Database, table and data loaded successfully')
```
